Судя по порядку величины и скоров в таблице, при валидации используется все таки RMSE, поэтому далее будем использовать именно его

```python
def corr(x, y):
    numerator = ((x - x.mean(axis=-1, keepdims=True))*(y - y.mean(axis=-1, keepdims=True))).sum(axis=-1)
    denominator = (np.sum((x - x.mean(axis=-1, keepdims=True))**2, axis=-1) * np.sum((y - y.mean(axis=-1, keepdims=True))**2, axis=-1)) ** 0.5
    return numerator / denominator
```

```python
from sklearn.decomposition import PCA

pca_std = PCA()
pca_mm = PCA()
pca_std.fit(StandardScaler().fit_transform(X))
pca_mm.fit(MinMaxScaler().fit_transform(X))

n_features = np.arange(X.shape[1])
explained_variance_std = pca_std.explained_variance_ratio_
explained_variance_mm = pca_mm.explained_variance_ratio_

plt.figure(figsize=(10, 8))

plt.bar(n_features, explained_variance_std, label='explained variance StandardScaler',
       color=sns.color_palette()[3])
plt.plot(n_features, np.cumsum(explained_variance_std), marker='o', label='cumsum StandardScaler',
       color=sns.color_palette()[3])
plt.bar(n_features, explained_variance_mm, label='explained variance MinMaxScaler',
       color=sns.color_palette()[0])
plt.plot(n_features, np.cumsum(explained_variance_mm), marker='o', label='cumsum MinMaxScaler',
       color=sns.color_palette()[0])

plt.xlabel("number of features", fontsize=18)
plt.ylabel("variance proportion", fontsize=18)
plt.title(f"PCA", fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(bottom=0, top=1.05)

plt.legend(shadow = True, fontsize=18, loc='lower right')

plt.show()
```

```python
batch_size = 500

train_dataset = TensorDataset(torch.Tensor(X_train), torch.Tensor(y_train))
eval_dataset = TensorDataset(torch.Tensor(X_eval), torch.Tensor(y_eval))

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
eval_loader = DataLoader(eval_dataset, batch_size=y_eval.shape[0], shuffle=False)
```

```python
model = nn.Sequential(
    nn.Linear(100,20),
    nn.SiLU(),
    nn.Linear(20,1)
).cuda()

n_epochs = 300

train_losses = []
eval_losses = []

optimizer = torch.optim.Adam(model.parameters(), lr=1e-2,
                             weight_decay=1e-8)
mse_loss = torch.nn.functional.mse_loss


for epoch in range(n_epochs):
    cur_losses = []
    for i, (x_tr,y_tr) in enumerate(train_loader):
        optimizer.zero_grad()
        x_tr, y_tr = x_tr.cuda(), y_tr.cuda()
        y_preds = model(x_tr)
        
        loss = torch.sqrt(mse_loss(y_preds, y_tr.reshape(-1,1)))
        loss.backward()
        optimizer.step()
        
        cur_losses.append(loss.detach().cpu())
    mean_loss = np.mean(cur_losses)
    train_losses.append(mean_loss)
    
    with torch.no_grad():
        val_pred = model(torch.Tensor(X_eval).cuda()).detach().cpu()
        val_loss = torch.sqrt(mse_loss(val_pred, torch.Tensor(y_eval).reshape(-1,1))).detach().cpu()
        eval_losses.append(val_loss)
    if (epoch+1)%40 == 0:
        clear_output(wait=True)
        plt.plot(train_losses, label="train")
        plt.plot(eval_losses, label="eval")
        plt.legend()
        plt.show()
print(np.min(eval_losses))
```

Попробуем что-то вроде MoE

```python
class Expert(nn.Module):
    def __init__(self, hidden_dim=20):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(100,hidden_dim),
                                 nn.SiLU(),
                                 nn.Linear(hidden_dim,1))
    def forward(self, inp):
        return self.net(inp)

class MoE_NN(nn.Module):
    def __init__(self, n_experts=4, weights_hidden_dim=10, experts_hidden_dim=20):
        super().__init__()
        self.n_experts = n_experts
        self.experts_weights = nn.Sequential(nn.Linear(100, weights_hidden_dim),
                                             nn.SiLU(),
                                             nn.Linear(weights_hidden_dim, n_experts),
                                             nn.Softmax(dim=-1)).cuda()
        self.experts = []
        for i in range(n_experts):
            self.experts.append(Expert(experts_hidden_dim).cuda())
        
    def forward(self, inp):
        weights = self.experts_weights(inp)
        experts_preds = torch.stack([expert(inp) for expert in self.experts], dim=-1)
        out = (experts_preds * weights.reshape(inp.shape[0], 1, self.n_experts)).sum(axis=-1)
        return out


n_epochs = 500

train_losses = []
eval_losses = []

MoE_model =  MoE_NN(n_experts=4, weights_hidden_dim=5, experts_hidden_dim=20)

optimizer = torch.optim.Adam(MoE_model.parameters(), lr=1e-5)
mse_loss = torch.nn.functional.mse_loss


for epoch in range(n_epochs):
    cur_losses = []
    for i, (x_tr,y_tr) in enumerate(train_loader):
        optimizer.zero_grad()
        x_tr, y_tr = x_tr.cuda(), y_tr.cuda()
        y_preds = torch.exp(MoE_model(x_tr))
        
        loss = torch.sqrt(mse_loss(y_preds, y_tr.reshape(-1,1)))
        loss.backward()
        optimizer.step()
        cur_losses.append(loss.detach().cpu())
        
    train_losses.append(np.mean(cur_losses))
    
    with torch.no_grad():
        val_pred = torch.exp(MoE_model(torch.Tensor(X_eval).cuda()).detach().cpu())
        val_loss = torch.sqrt(mse_loss(val_pred, torch.Tensor(y_eval).reshape(-1,1))).detach().cpu()
        eval_losses.append(val_loss)
    if (epoch+1)%10 == 0:
        clear_output(wait=True)
        plt.plot(train_losses, label="train")
        plt.plot(eval_losses, label="eval")
        plt.legend()
        plt.show()
print(np.min(eval_losses))
```

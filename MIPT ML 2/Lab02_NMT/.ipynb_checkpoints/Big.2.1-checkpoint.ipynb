В этой части задания вы будете работать с данными недавно завершившегося [соревнования](https://github.com/dialogue-evaluation/RuSentNE-evaluation) по анализу тональности к именованным сущностям в новостных текстах. Посмотрим на описание, которое представлено в репозитории. 

Анализ тональности текста &mdash; извлечение выраженной в тексте эмоциональной оценки к некоторой сущности &mdash; одно из наиболее активно развивающихся направлений в автоматической обработке текстов.

Анализ тональности новостных текстов &mdash; важное направление в области анализа мнений, поскольку обнаружение, отслеживание трендов тональности в новостном потоке важно для построения разного рода аналитических систем, отслеживания имиджа в СМИ конкретных людей или компаний.

Тональность по отношению к сущности в новостном тексте может происходить по крайней мере из трех разных источников:
* мнения автора текста;
* цитируемого мнения, при этом сам носитель мнения может быть упомянут или не упомянут в тексте;
* имплицитного мнения, которое следует из каких-либо упомянутых действий или реакций, например, X уволил Y. Такая информация часто присутствует даже при внешне нейтральном изложении событий.

Как мы видим, предложения длиннее 426 символов не встречаются, скорее всего, после токенизации это число только уменьшится. Выделим специальные токены, добавим ещё один, обозначающий конец группы, начатой другим специальным токеном, после этого инициализируем модель, указав ей наши специальные токены. Остановим здесь выбор на rubert-tiny2, так как он маленький, быстрый, а также показывает хорошие результаты.

````python
spec_tokens = train_df["entity_tag"].unique()
for i, tok in enumerate(spec_tokens):

    # Окружим спец. токены квадратными скобками, чтобы вероятность встретить их была ещё меньше.
    spec_tokens[i] = "[" + tok + "]"

spec_tokens = np.hstack([spec_tokens, ["[GREND]"]]) # токен конца группы
spec_tokens = spec_tokens.astype(str)
print(spec_tokens)

# Загружаем и настраиваем модель
tokenizer = AutoTokenizer.from_pretrained("cointegrated/rubert-tiny2")
bert_processor = AutoModelForTokenClassification.from_pretrained("cointegrated/rubert-tiny2",
                                                                 num_labels=512)
tokenizer.add_special_tokens({"additional_special_tokens": spec_tokens.tolist()})
bert_processor.resize_token_embeddings(len(tokenizer))
````

Предобработаем данные. Для этого напишем вспомогательную функцию для удобного кодирования данных.

````python
def encode(row):
    """Извлекаем из строки таблицы дынные в удобном формате"""
    
    # Выдаляем необъ. части предложения
    entity_start = row["entity_pos_start_rel"]
    entity_end = row["entity_pos_end_rel"]
    tag = "[" + row["entity_tag"] + "]"
    sent = row["sentence"][:entity_start] + tag + row["sentence"][entity_start:entity_end] + "[GREND]" + \
           row["sentence"][entity_end:]
    
    # Кодируем предложение
    tokenized_sent = tokenizer(
        sent,
        add_special_tokens=True,
        max_length=196, # Разумно урежем
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_tensors='pt',
    )
    
    tag_pos = tokenized_sent['input_ids'][0].tolist().index(tokenizer.convert_tokens_to_ids(tag))
    
    # Кодируем метку по аналогу какого-то задания с NLP
    label = [0, 0, 0]
    label[row['label'] + 1] = 1
    
    return tokenized_sent, label, tag_pos
````

Сделаем новый датасет.

````python
raw_data = train_df.apply(encode, axis=1)
train_tf_dataset = Dataset.from_dict({
    'input_ids': [x[0]['input_ids'].flatten() for x in raw_data],
    'attention_mask': [x[0]['attention_mask'].flatten() for x in raw_data],
    'token_idx': [x[2] for x in raw_data],
    'label': [x[1] for x in raw_data],
})


val_df = pd.read_csv("./validation_data_labeled.csv", sep="\t", header=0)
raw_data = val_df.apply(encode, axis=1)
val_tf_dataset = Dataset.from_dict({
    'input_ids': [x[0]['input_ids'].flatten() for x in raw_data],
    'attention_mask': [x[0]['attention_mask'].flatten() for x in raw_data],
    'token_idx': [x[2] for x in raw_data],
    'label': [x[1] for x in raw_data],
})
````

Найдём максимальное кол-во неенулевых символов в предложении и проверим, насколько адеквтано мы выбрали лимит длины текста.

````python
mm = 0
for text in train_tf_dataset["input_ids"]:
    m = np.count_nonzero(text)
    if m > mm:
        mm = m
mm
````

Как мы видим, размер адекватный с небольшой поправкрой на большие тексты на инференсе. Теперь напишем вспомогатльные функции для подсчёта наших метрик. Заодно напишем лосс в удобной для тренера форме.

````python
def ourF1(true, preds):
    scores = F1(true, preds, average=None, labels=[0, 1, 2], zero_division=0)
    return (scores[0] + scores[2]) / 2

def compute_metrics(eval_pred):
    """Считает метрику в удобном для тренера формате"""
    
    predictions = eval_pred.predictions
    labels = eval_pred.label_ids
    true = np.array(np.argmax(labels, axis=-1))
    preds = np.array(np.argmax(predictions, axis=-1))

    return {"F1": ourF1(true, preds)}
````

Напишем свой тренер для моделей.

````python
def collate_data(inp, device="cuda:0", has_label=True):
    """Обрабатывает данные для использования их моделью"""

    result = dict()
    result["input_ids"] = torch.tensor([x["input_ids"] for x in inp])
    result["attention_mask"] = torch.tensor([x["attention_mask"] for x in inp])
    result["token_idx"] = torch.tensor([x["token_idx"] for x in inp], device=device)
    if has_label:
        result["label"] = torch.tensor([x["label"] for x in inp], device=device, dtype=torch.float32)
    return result

# Наш тренер
class CustomTrainer(Trainer):
    """Тренер"""

    num_labels = 3

    # Функция подсчёта лосса
    def compute_loss(self, model, inputs, return_outputs=False):
        """Считает loss для нашей модели"""

        labels = inputs["label"]
        outputs = model.forward(**inputs)

        loss = nn.CrossEntropyLoss()(outputs.view(-1, self.num_labels), labels.view(-1, self.num_labels))
        return (loss, {'logits': outputs, 'loss': loss}) if return_outputs else loss
````

Напишем нашу модель и ещё одун попроще для сравнения результатов.

````python
class BertModel(nn.Module):
    """Модель на основе Bert"""

    def __init__(self, bert_processor, p=0.1):
        """Инициализируем модель"""

        # Инициализируем модель-родителя
        super().__init__()
        
        # Инициализируем новые модули
        self.bert = bert_processor
        self.ReLU = nn.ReLU()
        self.dropout = nn.Dropout(p)
        self.logitizer = nn.Linear(512, 3)
    
    def forward(self, input_ids, attention_mask, token_idx, **kwargs):
        """forward модели"""

        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        
        # Получим токены на позициях, соответствующих искомому токену
        logits = logits[torch.arange(logits.shape[0]), token_idx]
        logits = self.dropout(logits)
        logits = self.ReLU(logits)
        logits = self.logitizer(logits)
        return logits


# Напишем более простую модель для сравнения
class LSTMModel(nn.Module):
    """Прострая модель"""
    
    def __init__(self, emb_dict_size=83834, emb_dim=128, lstm_units=256, num_layers=2):
        """Инициализация модели"""

        # Инициализируем модель-родителя
        super().__init__()
        
        # Инициализируем новые модули
        self.embedding = nn.Embedding(emb_dict_size, emb_dim)
        self.lstm = nn.LSTM(emb_dim, lstm_units, batch_first=True, num_layers=num_layers, bidirectional=True)
        self.ReLU = nn.ReLU()
        self.logitizer = nn.Linear(2*lstm_units, 3)
        

    def forward(self, input_ids, attention_mask, token_idx, **kwargs):
        """forward модели"""

        tokens = self.embedding(input_ids)
        outputs, _ = self.lstm(tokens)
        outputs = outputs[torch.arange(outputs.shape[0]), token_idx]
        outputs = self.ReLU(outputs)
        outputs = self.logitizer(outputs)
        return outputs
````

Обучим модели.

````python
training_args = TrainingArguments(
    output_dir='results',
    num_train_epochs=15,
    per_device_train_batch_size=128,
    per_device_eval_batch_size=128,
    warmup_steps=400,
    weight_decay=0.01,
    logging_dir='logs',
    logging_steps=400,
    evaluation_strategy='steps',
    eval_steps=400,
    save_steps=400,
    save_total_limit=3,
    load_best_model_at_end=True,

    metric_for_best_model='eval_loss',
    
    include_inputs_for_metrics=True,
    label_names=['label'],
    
    dataloader_pin_memory=False
)
````

````python
device="cuda:0"
model1 = BertModel(bert_processor)
model1 = model1.to(device)

trainer = CustomTrainer(
    model=model1,
    args=training_args,
    train_dataset=train_tf_dataset,
    eval_dataset=val_tf_dataset, 
    compute_metrics=compute_metrics,
    data_collator=collate_data,
)

trainer.train()
````

Теперь обучим базовую модель.

````python
device="cuda:0"
model2 = LSTMModel()
model2 = model2.to(device)

trainer = CustomTrainer(
    model=model2,
    args=training_args,
    train_dataset=train_tf_dataset,
    eval_dataset=val_tf_dataset, 
    compute_metrics=compute_metrics,
    data_collator=collate_data,
)

trainer.train()
````

Напишем функцию для подсчёта рейтинга и применим её.

````python
def calculate_score(model):
    """Функция для удобного подсёте рейтинга моделей"""
    
    preds = []
    labels = []
    for item in val_tf_dataset:
        dict_params = collate_data([item])
        for k in dict_params.keys():
            dict_params[k] = dict_params[k].to(device)

        logits = model.forward(**dict_params)
        pred = (torch.argmax(logits, dim=-1) - 1).item()
        preds.append(pred)
        labels.append(np.argmax(item["label"]) - 1)
        
    score = F1(labels, preds, average='macro')
    return score
````

Посчитаем полученные рейтинги.

````python
print(f"Рейтинг модели на основе bert: {np.round(calculate_score(model1), 4)}")
print(f"Рейтинг модели на основе LSTM: {np.round(calculate_score(model2), 4)}")
````

````python
def encode2(row):
    """Извлекаем из строки таблицы дынные в удобном формате"""
    
    # Кодируем предложение
    tokenized_sent = tokenizer(
        row["sentence"],
        add_special_tokens=True,
        max_length=196, # Разумно урежем
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_tensors='pt',
    )
    
    # Кодируем метку по аналогу какого-то задания с NLP
    label = [0, 0, 0]
    label[row['label'] + 1] = 1
    
    return tokenized_sent, label
````

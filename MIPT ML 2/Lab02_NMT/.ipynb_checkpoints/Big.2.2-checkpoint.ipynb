Стратифицированно разделим датасет на тест и тренировку.

````python
splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.06, random_state=137)
for train, test in splitter.split(X=data_df["request"], y=data_df["class"], groups=data_df["class"]):
    train_df = data_df.iloc[train]
    test_df = data_df.iloc[test]

encoder = LabelEncoder()
train_data = train_df["request"]
train_labels = encoder.fit_transform(train_df["class"])

test_data = test_df["request"]
test_labels = encoder.transform(test_df["class"])
test_data, test_labels
````

Зададим гиперпараметры

````python
device = "cuda:0"
batch_size = 256
max_req_len = 128
````

В качестве модели будем использовать rubert-tiny2, так как она сильно меньше, чем большие модели, однако она выдаёт результаты не сильно хуже, а иногда и лучше, чем они.

````python
# Загрузка нужных моделей
tokenizer = AutoTokenizer.from_pretrained("cointegrated/rubert-tiny2")
bert_processor = AutoModel.from_pretrained("cointegrated/rubert-tiny2")

# Создлание датасетов
train_dataset = Dataset.from_dict({
    "text": train_data,     # запросы
    "labels": train_labels  # метки
})

val_dataset = Dataset.from_dict({
    "text": test_data,     # запросы
    "labels": test_labels  # метки
})

# Преобразование тренировочного датасета
train_dataset = train_dataset.map(  # применяем функцию к каждому элементу датасета
    lambda elem: tokenizer(
        elem["text"],               #  токенизируем поле text
        add_special_tokens=True,    #  автоматически добавляем CLS, SEP токены
        truncation=True,            #  обрезаем ли последовательность
        max_length=max_req_len              #  ограничение на длину 
                                    # (чуть-чуть порежем самые длинные тексты до степени двойки)
    )
)

# Преобразование валидационного датасета
val_dataset = val_dataset.map(    # применяем функцию к каждому элементу датасета
    lambda elem: tokenizer(
        elem["text"],             #  токенизируем поле text
        add_special_tokens=True,  #  автоматически добавляем CLS, SEP токены
        truncation=True,          #  обрезаем ли последовательность
        max_length=max_req_len            #  ограничение на длину
    )
)
````

Запишем архитектуру модели.

````python
class Model(nn.Module):
    def __init__(self, bert=bert_processor, dropout_p=0.3):
        """Инициализация модели"""
        
        super(Model, self).__init__()
        self.bert = bert.to(device)
        self.dropout = nn.Dropout(p=dropout_p)
        self.logitizer = nn.Linear(312, 142)
    
    def forward(self, input_ids, token_type_ids, attention_mask, **kwargs):
        """forward модели"""
        
        bert_out = self.bert.forward(input_ids=input_ids, token_type_ids=token_type_ids,
                                     attention_mask=attention_mask)
        cls_token = bert_out.last_hidden_state[:, 0]  # Во всех элементах батча берём нулевой токен
        dropped = self.dropout(cls_token)
        return self.logitizer(dropped)

model = Model().to(device)
````

Проверим кооректность работы модели.

````python
summary(model, input_data={"input_ids": dummy_input["input_ids"],
                           "token_type_ids": dummy_input["token_type_ids"],
                           "attention_mask": dummy_input["attention_mask"]})
````

Напишем функцию для представления данных в нужном нам виде.

````python
# Обработчик данных
def collate_data(inp, has_label=True):
    """Преобразует данные"""

    max_idx_len = 0
    for i in range(len(inp)):
        max_idx_len = max(max_idx_len, len(inp[i]['input_ids']))
        inp[i]['input_ids'] = torch.tensor(inp[i]['input_ids'], device=device)
        inp[i]['token_type_ids'] = torch.tensor(inp[i]['token_type_ids'], device=device)
        inp[i]['attention_mask'] = torch.tensor(inp[i]['attention_mask'], device=device)

    ret = {}
    ret['input_ids'] = torch.stack([torch.cat([x['input_ids'],
                                               torch.full((max_idx_len - len(x['input_ids']),),
                                                          tokenizer.pad_token_id,
                                                          device=device)]) for x in inp])
    ret['token_type_ids'] = torch.stack([torch.cat([x['token_type_ids'],
                                               torch.full((max_idx_len - len(x['token_type_ids']),),
                                                          0, device=device)]) for x in inp])
    ret['attention_mask'] = torch.stack([torch.cat([x['attention_mask'],
                                               torch.full((max_idx_len - len(x['attention_mask']),),
                                                          0, device=device)]) for x in inp])
    if has_label:
        ret['labels'] = torch.tensor([x['labels'] for x in inp], device=device, dtype=torch.long)
    return ret

# Инициализируем даталоадеры
train_loader = DataLoader(train_dataset,
                          batch_size=batch_size,
                          shuffle=True,
                          collate_fn=collate_data)

val_loader = DataLoader(val_dataset,
                        batch_size=batch_size,
                        shuffle=False,
                        collate_fn=collate_data)
````

Обучим модель.

````python
training_args = TrainingArguments(
    
    # Параметры обучения
    num_train_epochs=15,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    warmup_steps=400,
    weight_decay=0.01,
    evaluation_strategy='steps',
    
    # Параметры логирования
    output_dir='results',
    logging_dir='logs',
    logging_steps=100,
    eval_steps=100,
    save_steps=100,
    save_total_limit=3,
    load_best_model_at_end=True,

    # Другие параметры
    metric_for_best_model='eval_loss',
    include_inputs_for_metrics=True,
    label_names=['labels'],
    
    dataloader_pin_memory=False
)


# Кастомный тренер с переопределённой функцией потерь
class CustomTrainer(Trainer):
    num_labels = len(np.unique(train_labels))

    def compute_loss(self, model, inputs, return_outputs=False):
        """Считаем правильный лосс"""

        y = inputs["labels"]
        preds = model(**inputs)

        loss_func = nn.CrossEntropyLoss()
        loss = loss_func(preds, y)
        return (loss, {'logits':preds, 'loss':loss}) if return_outputs else loss

# Каастомный подсчёт метрик
def compute_metrics(model_predictions):
    """Считаем правильную метрику"""

    preds = model_predictions.predictions
    labels = model_predictions.label_ids
    
    true_vals = np.array(labels)
    pred_vals = np.array(np.argmax(preds, axis=-1))

    return {"acc": accuracy(true_vals.reshape(-1, 1), pred_vals.reshape(-1, 1))}

# Инициализируем тренера
trainer = CustomTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset, 
    compute_metrics=compute_metrics,
    data_collator=collate_data,
)
````

````python
trainer.train()
````

Создадим новую модель и заново обучим её.

Как мы видим, дообучать с простой головой простой методикой получилось хуже, чем обучать на полном датасете, однако не очень сильно. Теперьб попробуем сделать голову двуслойной и увеличить размерность выходного слоя у берта.

````python
bert_processor2 = AutoModelForTokenClassification.from_pretrained("cointegrated/rubert-tiny2", num_labels=1024)

class ModelMLP(nn.Module):
    def __init__(self, bert=bert_processor2, dropout_p=0.15):
        """Инициализация модели"""
        
        super(ModelMLP, self).__init__()
        self.bert = bert.to(device)
        self.dropout1 = nn.Dropout(p=dropout_p)
        self.linear1 = nn.Linear(1024, 256)
        self.dropout2 = nn.Dropout(p=dropout_p)
        self.ReLU = nn.PReLU(num_parameters=256)
        self.logitizer = nn.Linear(256, 142)
    
    def forward(self, input_ids, token_type_ids, attention_mask, **kwargs):
        """forward модели"""
        
        bert_out = self.bert.forward(input_ids=input_ids, token_type_ids=token_type_ids,
                                     attention_mask=attention_mask)
        cls_token = bert_out.logits[:, 0]  # Во всех элементах батча берём нулевой токен
        result = self.dropout1(cls_token)
        result = self.linear1(result)
        result = self.dropout2(result)
        result = self.ReLU(result)
        return self.logitizer(result)
````

Обучим такую модель.

````python
model3 = ModelMLP().to(device)

training_args.num_train_epochs = 18
# Инициализируем тренера
trainer = CustomTrainer(
    model=model3,
    args=training_args,
    train_dataset=base_dataset,
    eval_dataset=val_dataset, 
    compute_metrics=compute_metrics,
    data_collator=collate_data,
)

trainer.train()

cont_args.num_train_epochs = 30
model3.bert.requires_grad_(False)
model3.dropout1.requires_grad_(False)

trainer = CustomTrainer(
    model=model3,
    args=cont_args,
    train_dataset=add_dataset,
    eval_dataset=val_dataset, 
    compute_metrics=compute_metrics,
    data_collator=collate_data,
)

trainer.train()
````

Как мы видим, с двойным слоем точность достигается гораздо быстрее, однако дообучение заметно рушит все результаты. Попробуем теперь применять средний пулинг к выходу.

````python
bert_processor1 = AutoModelForTokenClassification.from_pretrained("cointegrated/rubert-tiny2", num_labels=312)

class ModelLinear(nn.Module):
    def __init__(self, bert=bert_processor1, dropout_p=0.1):
        """Инициализация модели"""
        
        super(ModelLinear, self).__init__()
        self.bert = bert.to(device)
        self.dropout1 = nn.Dropout(p=dropout_p)
        self.linear1 = nn.Linear(312, 142)
    
    def forward(self, input_ids, token_type_ids, attention_mask, **kwargs):
        """forward модели"""
        
        bert_out = self.bert.forward(input_ids=input_ids, token_type_ids=token_type_ids,
                                     attention_mask=attention_mask)
        cls_token = bert_out.logits.mean(dim=-2)
        result = self.dropout1(cls_token)
        result = self.linear1(result)
        return result
````

````python
model4 = ModelLinear().to(device)

training_args.num_train_epochs = 18
# Инициализируем тренера
trainer = CustomTrainer(
    model=model4,
    args=training_args,
    train_dataset=base_dataset,
    eval_dataset=val_dataset, 
    compute_metrics=compute_metrics,
    data_collator=collate_data,
)

trainer.train()

cont_args.num_train_epochs = 30
model4.bert.requires_grad_(False)
model4.dropout1.requires_grad_(False)

trainer = CustomTrainer(
    model=model4,
    args=cont_args,
    train_dataset=add_dataset,
    eval_dataset=val_dataset, 
    compute_metrics=compute_metrics,
    data_collator=collate_data,
)

trainer.train()
````

Как мы видимм, усреднение особо результатов не изменило.

**Выводы:** В итоге мы понимаем, что модель с двойной головой обучается быстрее и точнее всего, однако дообучать её нужно либо с очень маленькими коэффициентами, либо на полном датасете, а не только на добавленных данных. Остальные модели особо разницы между собой не показали.

Класс должен поддерживать основные математические операции (+, -, \*, /) за счет перегрузки соответствующих магических методов. Также он должен поддерживать возможность получить число в алгебраической и полярной форме. Допускается использование модуля `math`.

````python
import math

class ComplexNumber:
    def __init__(self, re=0, im=0, r=None, phi=None):
        self.re = re
        self.im = im
        if r is not None and phi is not None:
            self.re = r * math.cos(phi)
            self.im = r * math.sin(phi)
    
    def from_algebraic(re, im):
        return ComplexNumber(re=re, im=im)
    
    def from_polar(r, phi):
        return ComplexNumber(r=r, phi=phi)
    
    def __str__(self):
        im = round(self.im, 5) if round(self.im, 5) != int(self.im) else int(self.im)
        re = round(self.re, 5) if round(self.re, 5) != int(self.re) else int(self.re)
            
        if (im != 0) & (re != 0):
            return f"({re} {'+' if im >= 0 else '-'} {str(abs(im)) + '*' if im!=1 else ''}i)"
        elif im != 0:
            return f"{'-'if im < 0 else ''}{str(abs(im)) + '*' if im!=1 else ''}i"
        elif re != 0:
            return f"{re}" if re >= 0 else f"({re})"
        else:
            return "0"

    def __add__(self, other):
        return ComplexNumber(re = self.re + other.re, im = self.im + other.im)

    def __sub__(self, other):
        return ComplexNumber(re = self.re - other.re, im = self.im - other.im)

    def __mul__(self, other):
        return ComplexNumber(re = self.re * other.re - self.im * other.im, im = self.re * other.im + self.im * other.re)

    def __truediv__(self, other):
        denom = other.sqabsolute()
        
        return ComplexNumber(re = (self.re * other.re + self.im * other.im) / denom ,
                            im = (self.im * other.re - self.re * other.im) / denom)

    def sqabsolute(self):
        return self.re**2 + self.im**2

    def arg(self):
        return math.atan2(self.im, self.re)

    def to_polar(self):
        r = math.sqrt(self.sqabsolute())
        phi = self.arg()
        return r, phi

    def to_algebraic(self):
        return self.re, self.im

    
a01 = ComplexNumber.from_algebraic(0, 1)
b1pidiv4 = ComplexNumber.from_polar(1, math.pi/4)

print(f"{a01} + {b1pidiv4} = {a01 + b1pidiv4}")
print(f"{a01} - {b1pidiv4} = {a01 - b1pidiv4}")
print(f"{a01} * {b1pidiv4} = {a01 * b1pidiv4}")
print(f"{a01} / {b1pidiv4} = {a01 / b1pidiv4}\n")

a10 = ComplexNumber.from_polar(1, 0)
bm10 = ComplexNumber.from_algebraic(-1, 0)
print(f"{a10} + {bm10} = {a10 + bm10}")
print(f"{a10} - {bm10} = {a10 - bm10}")
print(f"{a10} * {bm10} = {a10 * bm10}")
print(f"{a10} / {bm10} = {a10 / bm10}\n")

a2dot5pidiv4 = ComplexNumber.from_polar(2.5, math.pi/4)
b4pidiv4 = ComplexNumber.from_polar(4, math.pi/4)
print(f"{a2dot5pidiv4} + {b4pidiv4} = {a2dot5pidiv4 + b4pidiv4}")
print(f"{a2dot5pidiv4} - {b4pidiv4} = {a2dot5pidiv4 - b4pidiv4}")
print(f"{a2dot5pidiv4} * {b4pidiv4} = {a2dot5pidiv4 * b4pidiv4}")
print(f"{a2dot5pidiv4} / {b4pidiv4} = {a2dot5pidiv4 / b4pidiv4}\n")



a05 = ComplexNumber.from_algebraic(0, 5)
b02 = ComplexNumber.from_algebraic(0, 2)
print(f"{a05} + {b02} = {a05 + b02}")
print(f"{a05} - {b02} = {a05 - b02}")
print(f"{a05} * {b02} = {a05 * b02}")
print(f"{a05} / {b02} = {a05 / b02}")
````

__Задание 3 (2 балла):__ Опишите класс для векторов в N-мерном пространстве. В качестве основы  используйте список значений координат вектора, задаваемый `list`. Обеспечьте поддержку следующих операций: сложение, вычитание (с созданием нового вектора-результата), скалярное произведение, косинус угла, евклидова норма. Все операции, которые можно перегрузить с помощью магических методов, должны быть реализованы именно через них. Класс должен производить проверку консистентности аргументов для каждой операции и в случаях ошибок выбрасывать исключение `ValueError` с исчерпывающим объяснением ошибки.

````python
import math
class Vector:
    def __init__(self, vector_values_list):
        pass
    
class Vector:
    def __init__(self, vector_values_list):
        if isinstance(vector_values_list, (float, int, bool)):
            self.values = [vector_values_list]
        else:
            self.values = list(vector_values_list)

    def __len__(self):
        return len(self.values)

    def __getitem__(self, i):
        if len(self) > i:
            return self.values[i]
        else:
            raise ValueError(f"Vector has less than {i+1} elements")

    def __setitem__(self, i, value):
        if len(self) > i:
            self.values[i] = value
        else:
            raise ValueError(f"Vector has less than {i+1} elements")
            
    def __str__(self):
        return f"({', '.join([str(elem) for elem in self.values])})"
        
    def __add__(self, other):
        if len(self) != len(other):
            raise ValueError("Vectors must be the same length")
        result = []
        for i in range(len(self)):
            result.append(self[i] + other[i])
        return Vector(result)

    def __sub__(self, other):
        if len(self) != len(other):
            raise ValueError("Vectors must be the same length")
        result = []
        for i in range(len(self)):
            result.append(self[i] - other[i])
        return Vector(result)

    def __mul__(self, other):
        if isinstance(other, (int, float, bool)):
            result = []
            for i in range(len(self)):
                result.append(self[i] * other)  
            return Vector(result)
        elif isinstance(other, Vector):
            if len(self) != len(other):
                raise ValueError("Vectors must have the same dimension")
            result = 0
            for i in range(len(self)):
                result += self[i] * other[i]
                return result
        else:
            raise ValueError("Multiplication is only defined between a vector and a scalar or another vector") 
    
    def __rmul__(self, other):
        if isinstance(other, (int, float, bool)):
            result = []
            for i in range(len(self)):
                result.append(self[i] * other)
            return Vector(result)
        else:
            raise ValueError("Multiplication is only defined between a vector and a scalar or another vector")    
            
    def __truediv__(self, other):
        if isinstance(other, (int, float, bool)):
            if other == 0: 
                raise ValueError("can't divide by zero")
            result = []
            for i in range(len(self)):
                result.append(self[i] / other)
        else:
            raise ValueError("Division is only defined between a vector and a scalar")
        return Vector(result)

    def __pos__(self):
        return self

    def __neg__(self):
        result = []
        for i in range(len(self)):
            result.append(-self[i])
        return Vector(result)
    
    def __eq__(self, other):
        if not isinstance(other, Vector):
            return False
        elif len(other) != len(self):
            return False
        else:
            for i in range(len(self)):
                if self[i] != other[i]:
                    return False
            return True

    def cos(self, other):
        if len(self) != len(other):
            raise ValueError("Vectors must have the same dimension")
        numerator = self * other
        denominator = self.norm() * other.norm()
        if denominator == 0:
            raise ValueError("One of the vectors is null")
        return numerator / denominator

    def norm(self):
        return math.sqrt(sum([x**2 for x in self.values]))
````

__Задание 4 (2 балл):__ Опишите декоратор, который принимает на вход функцию и при каждом её вызове печатает строку "This function was called N times", где N - число раз, которое это функция была вызвана на текущий момент (пока функция существует как объект, это число, очевидно, может только неубывать).

````python
def calls_counter(func):
    def wrapper(*args, **kwargs):
        wrapper.calls += 1
        print(f"This function was called {wrapper.calls} {'times' if wrapper.calls > 1 else 'time'}")
        return func(*args, **kwargs)
    wrapper.calls = 0
    return wrapper

@calls_counter
def foo(x):
    return x * 10

x = 1
print(foo(foo(foo(x))))
````

__Задание 5 (3 балла):__ Опишите декоратор класса, который принимает на вход другой класс и снабжает декорируемый класс всеми атрибутами входного класса, названия которых НЕ начинаются с "\_". В случае конфликтов имён импортируемый атрибут должен получить имя с суффиксом "\_new".

````python
def copy_class_attrs(cls):
    def wrapper(cls_upd, *args, **kwargs):
        for name in dir(cls):
            if not name.startswith("_"):
                attr = getattr(cls, name)
                if not hasattr(cls_upd, name):
                    setattr(cls_upd, name, attr)
                else:
                    setattr(cls_upd, f'{name}_new', attr)
        return cls_upd
    return wrapper

class clA:
    aaa = 1000
    def __init__(self, a=1, b=1):
        self.a = a
        self.b = b
    def sum_ab(self):
        print(self.a + self.b)

@copy_class_attrs(clA)
class clB:
    def __init__(self, a=0, b=10):
        self.a = a
        self.b = b

b_obj = clB(1, 5)

print(f"'sum_ab' in dir(b_obj): {'sum_ab' in dir(b_obj)}")
print("b_obj.sum_ab(): ", end='')
b_obj.sum_ab()
b_obj.aaa
````

В реализации математических операций запрещается пользоваться любыми функциями, требующими использования оператора `import`.

````python
class SaveMixin:
    def save_txt(self, file_path):
        with open(file_path, 'w') as f:
            serialized_obj = str(self.__dict__)
            f.write(serialized_obj)
            
    def save_binary(self, file_path):
        with open(file_path, 'wb') as f:
            serialized_obj = str(self.__dict__)
            f.write(serialized_obj.encode('utf-8'))
            

class LoadMixin:
    def load_txt(file_path, cls):
        with open(file_path, 'r') as f:
            content = f.read()
            obj_dict = eval(content)
            obj = cls(list(*obj_dict.values()))
        return obj
    
    def load_binary(file_path, cls):
        with open(file_path, 'rb') as f:
            content = f.read().decode('utf-8')
            obj_dict = eval(content)
            obj = cls(list(*obj_dict.values()))
        return obj
        

class CopyMixin:
    def copy(self):
        cls = self.__class__()
        for key, value in self.__dict__.items():
            setattr(new_obj, key, value)
        return new_obj
    
    
class Matrix(SaveMixin, LoadMixin, CopyMixin):
    def __init__(self, *args):
        if len(args) == 1:
            if isinstance(args[0], list) and all(isinstance(row, list) for row in args[0]):
                if len(set([len(row) for row in args[0]])) != 1:
                    raise ValueError("Invalid values, lists must be the same lengths")
                if not all([isinstance(num, (int, float)) for row in args[0] for num in row]):
                    raise TypeError("Ivalid values, lists must consist of numbers")       
                self._matrix = [[num for num in row] for row in args[0]]            
            else:
                if not all([isinstance(num, (int, float)) for num in args[0]]):
                    raise TypeError("Ivalid values, list must consist of numbers")
                self._matrix = [[num for num in args[0]]]
        elif len(args) == 2:
            self._matrix = [[0 for j in range(args[1])] for i in range(args[0])]
        elif len(args) == 3:
            self._matrix = [[args[2] for j in range(args[1])] for i in range(args[0])]
        else:
            raise ValueError("Invalid values, arguments must be a single list, list of lists, size (2 integers) or size with default value")
            
    def __getitem__(self, index):
        return self._matrix[index]
    
    def __setitem__(self, index, value):
        if isinstance(value, list):
            if len(value) != len(self._matrix[index]):
                raise ValueError("Invalid size for matrix")
                if not all([isintance(num, (int, float)) for num in value]):
                    raise ValueError("Invalid values, list must consist of numbers")
            self._matrix[index] = value
        elif isinstance(value, (int, float)):
            self._matrix[index] = [value] * len(self._matrix[index])
        else:
            raise ValueError("Invalid value, must be list or number")
            
    def __iter__(self):
        return iter(self._matrix)
    
    def iterrows(self):
        return self.__iter__
    
    def itercols(self):
        return self.transpose().__iter__
    
    def __len__(self):
        return len(self._matrix)
    
    def size(self):
        return len(self), len(self._matrix[0])
    
    def __str__(self):
        return self._matrix.__str__()
    
    def __add__(self, other):
        if isinstance(other, Matrix):
            if self.size() != other.size():
                raise ValueError("Matrices are not the same size")
            res = [[self[i][j] + other[i][j] for j in range(self.size()[1])] for i in range(self.size()[0])]
        elif isinstance(other, (int, float)):
            res = [[self[i][j] + other for j in range(self.size()[1])] for i in range(self.size()[0])]
        else:
            raise TypeError(f"Can't add {type(other)} to Matrix")
        return Matrix(res)
    
    def __radd__(self, other):
        if isinstance(other, (int, float)):
            res = [[self[i][j] + other for j in range(self.size()[1])] for i in range(self.size()[0])]
        else:
            raise TypeError(f"Can't add Matrix to {type(other)}")
        return Matrix(res)
    
    def __sub__(self, other):
        if isinstance(other, Matrix):
            if self.size() != other.size():
                raise ValueError("Matrices are not the same size")
            res = [[self[i][j] - other[i][j] for j in range(self.size()[1])] for i in range(self.size()[0])]
        elif isinstance(other, (int, float)):
            res = [[self[i][j] - other for j in range(self.size()[1])] for i in range(self.size()[0])]
        else:
            raise TypeError(f"Can't subtract {type(other)} from Matrix")
        return Matrix(res)
    
    def __rsub__(self, other):
        if isinstance(other, (int, float)):
            res = [[-self[i][j] + other for j in range(self.size()[1])] for i in range(self.size()[0])]
        else:
            raise TypeError(f"Can't subtract Matrix from {type(other)}")
        return Matrix(res)
    
    def __mul__(self, other):
        if isinstance(other, (int, float)):
            res = [[self[i][j] * other for j in range(self.size()[1])] for i in range(self.size()[0])]
        else:
            raise TypeError(f"Can't multiply Matrix by {type(other)}")
        return Matrix(res)
            
    def __rmul__(self, other):
        if isinstance(other, (int, float)):
            res = [[self[i][j] * other for j in range(self.size()[1])] for i in range(self.size()[0])]
        else:
            raise TypeError(f"Can't multiply Matrix by {type(other)}")
        return Matrix(res)
    
    def __truediv__(self, other):
        if isinstance(other, (int, float)):
            if other == 0: 
                raise ValueError("Can't divide by zero")
            res = [[self[i][j] / other for j in range(self.size()[1])] for i in range(self.size()[0])]
        else:
            raise TypeError(f"Can't divide Matrix by {type(other)}")
        return Matrix(res)
    
    def dot(self, other):
        if isinstance(other, (int, float)):
            return self * other 
        elif isinstance(other, Matrix):
            if self.size()[1] != other.size()[0]:
                raise ValueError(f"Can't multiply matrices with sizes ({self.size()}) and ({other.size()})")
            res = [[sum([self[i][k]*other[k][j] for k in range(self.size()[1])]) for j in range(other.size()[1])] for i in range(self.size()[0])]
        else:
            raise TypeError(f"Can't multiply Matrix to {type(other)}")
        return Matrix(res)
    
    def __matmul__(self, other):
        if not isinstance(other, Matrix):
            raise TypeError(f"Can't matmul Matrix to {type(other)}")
        return self.dot(other)
    
    def __eq__(self, other):
        if not isinstance(other, Matrix):
            return False
        if not self.size() == other.size():
            return False
        if not self._matrix == other._matrix:
            return False
        return True 
````

````python
class Coffee:
    def __init__(self):
        self.cost = 0
    
    def calculate_cost(self):
        return self.cost

    
class Latte(Coffee):
    def __init__(self, other=None):
        if isinstance(other, Coffee):
            raise TypeError("There must be exactly one type of coffee in a cup")
        elif other is not None:
            raise TypeError(f"Incorrect argument class {other.__class__.__name__}")
        super().__init__()
        self.cost = 150


class Cappuccino(Coffee):
    def __init__(self, other=None):
        if isinstance(other, Coffee):
            raise TypeError("There must be exactly one type of coffee in a cup")
        elif other is not None:
            raise TypeError(f"Incorrect argument class {other.__class__.__name__}")
        super().__init__()
        self.cost = 140


class Americano(Coffee):
    def __init__(self, other=None):
        if isinstance(other, Coffee):
            raise TypeError("There must be exactly one type of coffee in a cup")
        elif other is not None:
            raise TypeError(f"Incorrect argument class {other.__class__.__name__}")
        super().__init__()
        self.cost = 100


class Add(Coffee):
    def __init__(self, coffee=None):
        if isinstance(coffee, Coffee):
            super().__init__()
            self.coffee = coffee
        else:
            raise TypeError(f"Can't add {self.__class__.__name__} to {coffee.__class__.__name__}")
    
    def calculate_cost(self):
        return self.coffee.calculate_cost() + self.cost


class Sugar(Add):
    def __init__(self, coffee=None):
        super().__init__(coffee)
        self.cost = 10


class Cream(Add):
    def __init__(self, coffee=None):
        super().__init__(coffee)
        self.cost = 30


class Cardamom(Add):
    def __init__(self, coffee=None):
        super().__init__(coffee)
        self.cost = 20

        
class Foam(Add):
    def __init__(self, coffee=None):
        super().__init__(coffee)
        self.cost = 20


class Syrup(Add):
    def __init__(self, coffee=None):
        super().__init__(coffee)
        self.cost = 20
````

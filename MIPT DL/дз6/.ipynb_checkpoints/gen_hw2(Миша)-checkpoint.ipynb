### Задача 2
Будем исследовать модели для трансфера стиля.

Как вы помните из лекции, самой первой моделью в данной области генеративного ИИ является [CycleGAN](https://arxiv.org/pdf/1703.10593v7.pdf) (2017). Однако прогресс не стоял на месте, и уже в следующем году появилась многообещающая модель [MUNIT](https://openaccess.thecvf.com/content_ECCV_2018/papers/Xun_Huang_Multimodal_Unsupervised_Image-to-image_ECCV_2018_paper.pdf) (2018). Авторы данной модели внедрили в CуcleGAN идею StyleGAN о возможности кодирования стиля. 

Напомним, что в CycleGAN использовалось две пары генераторов для трансфера изображения из первого домена во второй и из второго в первый соответственно, а также два дискриминатора для первого и второго доменов. В MUNIT также используется по две пары генераторов и дискриминаторов. Основное отличие состоит в том, они добавили в архитектуру модели часть с кодированием стиля.  Для этого они разделили каждый генератор на две части: кодировщик $\mathcal{E}$ и декодировщик $\mathcal{G}$.

В свою очередь кодировщик $\mathcal{E}$ также состоит из двух частей $\mathcal{E}^c$ и $\mathcal{E}^s$, которые преобразует входное изображение в латентные вектора контента и стиля соответственно.

Декодировщик $\mathcal{G}$ принимает на вход вектор контента, и с помощью вектора стиля преобразует его в новое изображение. На схеме ниже представлена схема генератора.



image.png

Кроме того авторы добавили новые лоссы для эффективного обучения модели:
* $\mathcal{L}_{recon}^{x}(\mathcal{E}_1, \mathcal{G}_1)=\mathsf{E}\big\|G(\mathcal{E}_1^c(X),\mathcal{E}_1^s(X))-X_1\big\|_1$ &mdash; лосс реконструкции изображения.
* $\mathcal{L}_{recon}^{x}(\mathcal{E}_2, \mathcal{G}_2)=\mathsf{E}\big\|G(\mathcal{E}_2^c(X),\mathcal{E}_2^s(X))-X_2\big\|_1$ &mdash; лосс реконструкции изображения.
*
* $\mathcal{L}_{recon}^{c}(\mathcal{E}_1, \mathcal{E}_2, \mathcal{G}_2)=\mathsf{E}\left\|\mathcal{E}_2^c(\mathcal{G}_2(C_1, S_2)) - C_1\right\|_1$ &mdash; лосс реконстукции латентного вектора контента, где $C_1=\mathcal{E}_1^c(X_1)$.
* $\mathcal{L}_{recon}^{c}(\mathcal{E}_2, \mathcal{E}_1, \mathcal{G}_1)=\mathsf{E}\left\|\mathcal{E}_1^c(\mathcal{G}_1(C_2, S_1)) - C_2\right\|_1$ &mdash; лосс реконстукции латентного вектора контента, где $C_2=\mathcal{E}_2^c(X_2)$.
*
* $\mathcal{L}_{recon}^{s}(\mathcal{E}_1, \mathcal{E}_2, \mathcal{G}_2)=\mathsf{E}\left\|\mathcal{E}_2^c(\mathcal{G}_2(C_1, S_2)) - S_2\right\|_1$ &mdash; лосс реконстукции латентного вектора стиля, где $C_1=\mathcal{E}_1^c(X_1)$.
* $\mathcal{L}_{recon}^{s}(\mathcal{E}_2, \mathcal{E}_1, \mathcal{G}_1)=\mathsf{E}\left\|\mathcal{E}_1^c(\mathcal{G}_1(C_2, S_1)) - S_1\right\|_1$ &mdash; лосс реконстукции латентного вектора стиля, где $C_2=\mathcal{E}_2^c(X_2)$.
*
* $\mathcal{L}_{GAN}^{x}(\mathcal{E}_1, \mathcal{G}_2, \mathcal{D}_2) = \mathsf{E}\log(1-\mathcal{D}_2(\mathcal{G}_2(C_1, S_2))) + \mathsf{E}\log \mathcal{D}_2(X_2)$ &mdash; адверсальный лосс, где $C_1=\mathcal{E}_1^c(X_1)$.
* $\mathcal{L}_{GAN}^{x}(\mathcal{E}_2, \mathcal{G}_1, \mathcal{D}_1) = \mathsf{E}\log(1-\mathcal{D}_1(\mathcal{G}_1(C_2, S_1))) + \mathsf{E}\log \mathcal{D}_1(X_1)$ &mdash; адверсальный лосс, где $C_2=\mathcal{E}_2^c(X_2)$.

где
* $X_1, X_2$ &mdash; независимые случайные изображения из истинного распределения первого и второго доменов изображений соответственно.
* $S_1$, $S_2$ &mdash; вектора стиля, которые задаются как случайные вектора из $\mathcal{N}(0, I)$.

Задача оптимизации представляется как

$$\min_{\mathcal{E}_1,\mathcal{E}_2,\mathcal{G}_1,\mathcal{G}_2}\:\max_{\mathcal{D}_1, \mathcal{D}_2}\:\mathcal{L}(\mathcal{E}_1, \mathcal{E}_2, \mathcal{G}_1, \mathcal{G}_2, \mathcal{D}_1, \mathcal{D}_2) = $$
$$ = \mathcal{L}_{GAN}^{x}(\mathcal{E}_1, \mathcal{G}_2, \mathcal{D}_2) + \mathcal{L}_{GAN}^{x}(\mathcal{E}_2, \mathcal{G}_1, \mathcal{D}_1) + $$
$$ + \lambda_x(\mathcal{L}_{recon}^{x}(\mathcal{E}_1, \mathcal{G}_1) + \mathcal{L}_{recon}^{x}(\mathcal{E}_2, \mathcal{G}_2)) + $$
$$ + \lambda_c(\mathcal{L}_{recon}^{c}(\mathcal{E}_1, \mathcal{E}_2, \mathcal{G}_2) + \mathcal{L}_{recon}^{c}(\mathcal{E}_2, \mathcal{E}_1, \mathcal{G}_1)) + $$
$$ + \lambda_s(\mathcal{L}_{recon}^{s}(\mathcal{E}_1, \mathcal{E}_2, \mathcal{G}_2) + \mathcal{L}_{recon}^{s}(\mathcal{E}_2, \mathcal{E}_1, \mathcal{G}_1))$$

Ниже представлена схема для используемых в модели лоссов.
